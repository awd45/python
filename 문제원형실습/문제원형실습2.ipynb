```python
## 패키지 불러오기
import pandas as pd
from sqlalchemy import create_engine

#DB커넥션 열기
engine = create_engine('oracle+cx_oracle://kopo:kopo@192.168.110.112:1521/orcl')

# DB 테이블을 읽어 Data Frame 변수에 저장하기
productVolume = pd.read_sql_query('SELECT regionid,product,substr(yearweek,5,2)as week , avg(qty) as qty FROM kopo_channel_seasonality_new where substr(yearweek,5,2) <=52 group by regionid,product,substr(yearweek,5,2) order by regionid,product,substr(yearweek,5,2) asc' , engine)

productVolume
```

```python
## 인덱스값 재 정렬
sortedData = productVolume.reset_index(drop=True)
sortedData
```

```python
import numpy as np
sortedData["qty"] = np.where(sortedData["qty"] > 700000 , 700000 , sortedData["qty"])
sortedData
```

# 이동평균 구하기

```python
def sub_function(data):
    data["qty_ma"] = data["qty"].rolling(window = 17, center = True , min_periods=1).mean()
    return data
```

```python
sortedData2 = sortedData.groupby(["regionid","product"]).apply(sub_function)
sortedData2
```

# 그룹바이 함수 디버깅

```python
groups = sortedData2.groupby(["regionid","product"])
```

```python
data = groups.get_group(list(groups.groups)[0])
data.head()
```

# 변동률 구하기

```python
def sub_function1(data):
    data["ma_ratio"] = data["qty_ma"].rolling(window = 5, center = True , min_periods=1).std()
    return data
```

```python
sortedData2 = sortedData2.groupby(["regionid","product"]).apply(sub_function1)
sortedData2
```

```python
sortedData2["upper_bound"] = sortedData2["qty_ma"]+sortedData2["ma_ratio"]
sortedData2
```

```python
sortedData2["lower_bound"] = sortedData2["qty_ma"]-sortedData2["ma_ratio"]
sortedData2
```

```python
sortedData2["qty_purify"] = np.\
    where(sortedData2["qty"] > sortedData2["upper_bound"] , sortedData2["upper_bound"] , np.\
    where(sortedData2["qty"] < sortedData2["lower_bound"] ,sortedData2["lower_bound"] , sortedData2["qty"]))
```

```python
sortedData2.head()
```

# 스무딩처리

```python
def sub_function2(data):
    data["smoothing"] = data["qty_purify"].rolling(window = 5, center = True , min_periods=1).mean()
    return data
```

```python
sortedData2 = sortedData2.groupby(["regionid","product"]).apply(sub_function2)
sortedData2
```

```python
sortedData2["seasonality"] = sortedData2["qty"] / sortedData2["smoothing"]
sortedData2
```

```python
## 컬럼 대문자로 변경
sortedData2.columns = [x.upper() for x in sortedData2.columns]

#CSV 파일로 저장
sortedData2.to_csv("../dataset/result3.csv", index=False)
sortedData2.head()
```
