```python
import pandas as pd
from sqlalchemy import create_engine 
 
# DB 커넥션 열기
engine = create_engine('oracle+cx_oracle://kopo:kopo@192.168.110.112:1521/orcl') 

# DB 테이블을 읽어 Data Frame 변수에 저장하기
customerData = pd.read_sql_query('SELECT * FROM kopo_product_volume', engine) 


# 데이터 VIEW
customerData.head()
```

```python
# 컬럼을 대문자로 변경
customerData.columns=[x.upper() for x in customerData.columns]
customerData.head()
```

```python
# 정렬해준다.
sortedDataPre = customerData.sort_values(["REGIONID", "PRODUCTGROUP", "YEARWEEK"])
```

```python
sortedDataPre
```

```python
# 어리럽혀진 인덱스를 리셋을 시켜준다.
sortedData = sortedDataPre.reset_index(drop=True)
sortedData
```

```python
# 이동편규 연산
import math
# 이동편균 구간
order = 5

# 앞뒤로 비어있게될 구간 크기
suborder = math.floor(order/2)
```

```python
suborder
```

```python
sortedData["MA"] = sortedData["VOLUME"].rolling(window = order, center = True).mean()
sortedData
```

```python
# 이동평균 구간 찾아보기
maxLength = len(sortedData)
maxLength
```

```python
# i가 0일때 선택범위
i=0
```

```python
sortedData["VOLUME"][(maxLength-1)-(suborder+i):maxLength]
```

```python
# # i가 1일때 선택범위
i=1
```

```python
sortedData["VOLUME"][(maxLength-1)-(suborder+i):maxLength]
```

```python
sortedData["VOLUME"][0+i:suborder]
```

```python
# 이동편균 뒤의 빈셀 채우기
postList=[]
maxLength = len(sortedData)
for i in range(0,suborder):
    postList.append(sortedData["VOLUME"][(maxLength-1)-(suborder+i):maxLength].mean())
    sortedData.loc[maxLength-(1+i),"MA"] = postList[i]
postList
```

```python
# 이동편균 얖의 빈셀 채우기
preList = []
for i in range(0,suborder):
    preList.append(sortedData["VOLUME"][0+i:suborder].mean())
    sortedData.loc[i, "MA"] = preList[i]
preList
```

```python
sortedData
```

```python
# 다른 방식으로 디버깅
# 로케이션을 쓸때는 -가 안먹힘
sortedData.loc[-(2+suborder+1):maxLength]["VOLUME"]
```

```python
sortedData["VOLUME"][-(1+suborder+1):maxLength]
```

```python
sortedData["VOLUME"][-(0+suborder+1):maxLength]
```

```python
# 이동편균 뒤의 빈셀 채우기
postList=[]
maxLength = len(sortedData)
for i in range(0,suborder):
    postList.append(sortedData["VOLUME"][-(i+suborder+1):maxLength].mean())
    sortedData.loc[maxLength-(1+i),"MA"] = postList[i]
postList
```

```python
sortedData.tail()
```

```python
# 이동편균 얖의 빈셀 채우기
preList = []
for i in range(0,suborder):
    preList.append(sortedData["VOLUME"][0:suborder+i+1].mean())
    sortedData.loc[i, "MA"] = preList[i]
preList
```
