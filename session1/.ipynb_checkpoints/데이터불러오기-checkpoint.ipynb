### quiz 퀴즈내 링크만 전부 FOR LOOP를 활용하여 추출하세요!

```python
import requests, bs4, pandas as pd, numpy as np
 
#웹페이지 html 소스 가져오기
resp = requests.get('http://13.209.33.89:3333/quiz')
resp.encoding='utf-8'
html = resp.text
bs = bs4.BeautifulSoup(html, 'html.parser')
# html 부분구조 가져오기
a_links = bs.findAll("a")
for a in a_links:
    href= a.attrs["href"]
    print(href)

```

```python
import requests, bs4, pandas as pd, numpy as np
 
#웹페이지 html 소스 가져오기
resp = requests.get('http://13.209.33.89:3333/quiz')
resp.encoding='utf-8'
html = resp.text
bs = bs4.BeautifulSoup(html, 'html.parser')
# html 부분구조 가져오기
a_links = bs.findAll("a")
for a in a_links:
    #a 링크 내 href 속성
    href= a.attrs["href"]
    #a 링크 텍스트
    text = a.text
    if(href.startswith("/")):
        continue;
    print(text, ">", href)
```
