```python
# pandas 패키지 불러오기
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

#CSV 파일을 읽어 DATA Frame 변수에 저장하기
flowerData = pd.read_csv("../dataset/IRIS.csv")
print(flowerData.head(2))

species = flowerData[['SPECIES']]
feature = flowerData[['PETALLENGTH','PETALWIDTH']]
feature.head()

#create model and prediction
model = KMeans(n_clusters=3,algorithm='auto')
model.fit(feature)
predict = pd.DataFrame(model.predict(feature))
predict.columns=['PREDICT']
## 기본 컬럼은 인덱스번호로 되어있기때문에 컬럼값을 PREDICT로 바꿔준것

#conatenate labels to df as a new column
r = pd.concat([species,feature,predict],axis=1)
## SPECIES컬럼과 PETAL컬럼들과 PREDICT컬럼을 하나의 엑셀데이타로 붙이겠다.
# 결과를 CSV로 저장하는 방법 
r.to_csv("d:/iris_result.csv") 

#scatter plot 결과 데이터 시각화, C는 컬러를 의미한다.
plt.scatter(r['PETALLENGTH'],r['PETALWIDTH'],c=r['PREDICT'],alpha=1)

#pit.scatter(center_x,center_y,s=50,marker='0',c='r')
plt.show()

```
