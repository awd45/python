```python
#필요 라이브러리 정의
import numpy as np
import matplotlib.pyplot as plt

#데이터 정의 첫번째 컬럼은 정규분포, 두번째 컬럼은 *100
y=np.random.standard_normal((20,2))
y[:,1] = y[:,1]*100

plt.figure(figsize=(7,5))
plt.subplot(121) # 1행 2열 차트에서 두번째
plt.plot(y[:, 0], 'b',lw=1.5, label = '1st')
plt.plot(y[:, 0], 'o') # 점은 따로 그리고 중첩시켜 그래프와 같이 그려진것처럼 보여진다.
plt.grid(True)
plt.legend(loc=0)
plt.ylabel('value')
plt.title('A Simple Plot')
plt.subplot(122) # 1행 2열 차트에서 두번째
plt.plot(y[:, 1], 'g', lw=1.5, label='2nd')
plt.plot(y[:, 1], 'ro')
plt.grid(True)
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
```

```python
#필요 라이브러리 정의
import numpy as np
import matplotlib.pyplot as plt

#데이터 정의 첫번째 컬럼은 정규분포, 두번째 컬럼은 *100
y=np.random.standard_normal((20,2))
y[:,1] = y[:,1]*100 # 가중치를 곱해준다.

#첫번째 컬럼자료를 첫번째(왼쪽) 축에 대응하여 그림
fig, ax1 = plt.subplots()
plt.plot(y[:, 0], 'b', lw=1.5, label = '1st')
plt.plot(y[:, 0], 'bo')
plt.grid(True)
plt.legend(loc=4)
plt.xlabel('index')
plt.ylabel('value 1st')
plt.title('A Simple Plot')
#두번째 컬럼자료를 두번쨰(오른쪽) 축에 대응하여 그림
ax2 = ax1.twinx() # 한차트안에서 같이 그리는 함수
plt.plot(y[:, 1], 'g', lw=1.5, label = '2nd')
plt.plot(y[:, 1], 'go')
plt.legend(loc=0)
plt.ylabel('value 2nd')
```

```python
#pip install seasborn
import seaborn as sns
%matplotlib inline

tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.distplot(tips['total_bill'])
```

```python
#Jointplot 차트 구현 데이터셋에 대하여 특정 2가지 값에 대하여 그래프형태로 표현

# 특정두가지 값을 가지고 어느지점에 분포도까지 확인가능
```

```python
sns.jointplot(x= 'total_bill', y = 'tip', data = tips)
```

```python
#pairplot 차트 구현 전체 데이터의 모든관계를 한눈에 보기위한 차트

```

```python
sns.pairplot(tips)

sns.pairplot(tips, hue = 'sex')
```

```python
# barplot 차트 구현 카테고리 데이터에 대하여 구분하는 차트

```

```python
sns.barplot(x='sex', y='total_bill', data = tips)
```

```python
#countplot 차트 구현 해당개수를 세어주는 countplot
```

```python
sns.countplot(x= 'sex', data = tips)
```

```python
# boxplot 차트 구현 정해진 카테고리 기준에 어떤 범위의 분포를 보여주는지 차트화함
```

```python
sns.boxplot(x='day', y='total_bill',data=tips)
```

```python
# corr 데이터의 상관관계를 확인
# 코릴레이션 파악가능
```

```python
tips.corr()
```
